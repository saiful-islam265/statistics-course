```python
import matplotlib.pyplot as plt

# Data to plot
labels = 'Agriculture', 'Health', 'Education', 'Transport', 'Others'
sizes = [68.27, 58.34, 28.08, 18.36, 55.38]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'purple']
explode = (0.1, 0, 0, 0, 0)  # explode 1st slice (Agriculture)

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.title('Yearly Budget Distribution by Sector')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to plot
sectors = ['Agriculture', 'Health', 'Education', 'Transport', 'Others']
expenditure = [68.27, 58.34, 28.08, 18.36, 55.38]

plt.bar(sectors, expenditure, color=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'purple'])
plt.xlabel('Sector')
plt.ylabel('Expenditure (Taka)')
plt.title('Yearly Budget by Sector')
plt.show()
```
